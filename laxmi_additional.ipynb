```python
# Q.1: Box plot for VO2 max by Sport
# Box plot for VO2 max by Gender


import seaborn as sns
Athletes = pd.read_excel('V02Max_Combined.xlsx','Athletes')
# Box plot for VO2 max by Sport
plt.figure(figsize=(10, 6))
sns.boxplot(x='sport', y='VO2 max (ml/kg)', data=Athletes)
plt.title('VO2 Max by Sport')
plt.show()

# Box plot for VO2 max by Gender
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='VO2 max (ml/kg)', data=Athletes)
plt.title('VO2 Max by Gender')
plt.show()
```

```python
# Q.2: Create a correlation matrix to show the relationship between weight, height and V02 max.
import pandas as pd

Athletes = pd.read_excel('V02Max_Combined.xlsx','Athletes')
correlation_matrix = Athletes[['Weight', 'Height', 'VO2 max (ml/kg)']].corr()
print(correlation_matrix)

# Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Q.3: calculate the average (mean) VO2 max (ml/kg) for each sport category 
Athletes = pd.read_excel('V02Max_Combined.xlsx','Athletes')

# Mean VO2 max by sport
sport = (Athletes.groupby('sport')['VO2 max (ml/kg)'].mean())
print(sport)

```
